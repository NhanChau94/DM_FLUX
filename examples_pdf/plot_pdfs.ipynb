```python
import pickle as pkl
import matplotlib.pyplot as plt
import numpy as np
```

```python
# Access stored pdf:
with open('./pdf2d.pkl', 'rb') as f:
    pdf = pkl.load(f)
f.close()
# pdf stores bin count in histogram as ndarray 
# in which RA is along the first dimention, sin Dec along second dimension
background = pdf['background_pdf_2d']
signal = pdf['signal_pdf_2d'] # in fact this is tautau, annihilation, NFW channel with mass 10 TeV in DNN cascades

# bin edges
ra_edges = pdf['RA_bin']
sin_dec_edges = pdf['sinDec_bin']
```

```python
# plotting signal PDF
X, Y = np.meshgrid(ra_edges[:-1], sin_dec_edges[:-1])
c = plt.pcolormesh(X, Y, signal.T, cmap='magma', shading='auto')
plt.colorbar(c, label='evt. fraction')
plt.xlabel('reco. right ascension [deg]')
plt.ylabel(r'$\sin \delta_{reco}$')
plt.title('Signal PDF')
# galactic center position:
plt.scatter(266.4, np.sin(np.deg2rad(-29.0)), color='red', label='Galactic center', marker='x')
```

```python
# plotting the 2 histograms
X, Y = np.meshgrid(ra_edges[:-1], sin_dec_edges[:-1])
c = plt.pcolormesh(X, Y, background.T, cmap='magma', shading='auto')
plt.colorbar(c, label='evt. fraction')
plt.xlabel('reco. right ascension [deg]')
plt.ylabel(r'$\sin \delta_{reco}$')
plt.title('Background PDF')
# galactic center position:
plt.scatter(266.4, np.sin(np.deg2rad(-29.0)), color='red', label='Galactic center', marker='x')
```
